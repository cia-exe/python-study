```python
def fib(n):
  """Print a Fibonacci series up to n.""" 
  # The 1st str is a docstring that can produce doc or let users browse the code.
  
  print('Fibonacci: ', end='')
  a, b = 0, 1
  c, d = 0, 1
  while a < n:
      print(a , end=',') # The argument end can be used to avoid the newline after the output
      if a!=c: raise Exception(a, "!=", c)
      a, b = b, a+b   # b, a = a+b, b # works too
      t=c; c=d; d=c+t # it works too.
  print() # print \n
  # functions without a return statement do return a value 'None'

print('docstring=', fib.__doc__)  # show docstring
fib(3333)
print(fib(3333))  # it will print 'None' because fib returns 'None'

print('\n----- dict -----')
users = {'Hans': 1, 'Éléonore': 0, '景太郎': 1} #dict
tmp = users # a reference, not a copy.
users['景太'] = 1
users['XXX'] = 3
del tmp['XXX']
print('dict reference assign test', tmp, users, tmp==users,'---', sep=' ;\n') # use 'sep' to change separator
for user, status in users.items(): 
  if status != 0:
     print('(',user,status, end=') ')
print('ok')

print('\n----- tuple & dict -----')
tup = 1,'2',[3,4,5],(6,7,8,9,0),{'a','b','c'} # A tuple is a immutable list.
print(type(tup), tup)

users = ('Hans',1), ('Éléonore',0), ('景太',1), ('吉胖喵',0), ('小石獅',0) # It can omit the outermost parentheses
res1 = dict(users)
res2 = []
for user,status in users:
    if status == 0:
      del res1[user]
    else:
      res2.append((user,status))

result1 = list(res1.items()) # convert dict to list
print(users, result1, res2, result1==res2, sep='\n')

print('\n----- range & map -----')
# class range(start, stop[, step=1])
# class range(stop) -> range(0, stop)
rgs = [range(5),range(5,10),range(0, 10, 3),range(-11, -100, -22)]
rg_lst = tuple(map(lambda r,u:(u[0], r, tuple(r), sum(r)), rgs, users)) # merge users and ranges to tuple by map()
print(*rg_lst, sep='\n') # unpack lists by '*'
print('ranges=',len(rgs),', users=',len(users),', result=',len(rg_lst)) # len(rg_lst) = min(len(rgs), len(users))

```

```python
import sys

# Exception hierarchy: https://docs.python.org/3/library/exceptions.html#exception-hierarchy
# inherit from Exception, as is officially recommended 
# (https://docs.python.org/3/tutorial/errors.html#user-defined-exceptions)

class B(Exception):
    pass

class C(B):
    pass

class D(C):
    pass

print('info of B:', type(B), B.__class__,'@base', B.__base__,
      '\n@dict', D.__dict__,'\n@ctx', D.__context__)

print('\n------  basic except ------')
for cls in [B, C, D, TypeError, NameError, BaseException]:
    try:
        raise cls(1,2,3)
    except (C,D) as e: print("CD", e.__class__, e.args)
    #except B: print("B")
    except Exception as e: print("except:", e.__class__, e.args)
    except BaseException as e: print("baseExcept:", e.__class__, e.args)
    except: # omit the name to serve as a wildcard.
        print("omit name:", sys.exc_info()[:2])
        #raise # re-throw

print('\n------  except with else clause, ------')
try: 100/1
except: print('except')
else: print('else: no exception raised...')
finally: print('finally...')

print('\n------ OldStyleException ------')
class OldStyleException: pass

try:
   raise OldStyleException
#except Exception: print('Exception')          # can catch it
#except BaseException: print('BaseException')  # can catch it
except: print('Omit')                          # can catch it

print('\n------ auto clean up ------')
import os

fn1='f1.txt'
fn2='f2.txt'

def remove(f):
  if os.path.exists(f):
     os.remove(f)

remove(fn1)
remove(fn2)

with open(fn1,'w') as f1, open(fn2,'w') as f2:
  for i,(l1,l2) in enumerate(zip(('aaa','bbb','ccc'),('dd','ee','ff'))):
    f1.write(f'{i}:{l1}+{l2}\n');
    f2.write(f'{i}:{l2}+{l1}\n');

with open(fn1) as f1, open(fn2) as f2:  # the default file mode is 'r'
  for l1,l2 in zip(f1,f2):
    print(l1[:-1],l2[:-1])
```
